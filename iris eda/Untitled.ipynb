```python
import numpy as np
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
data=pd.read_csv("haberman.csv",names=['age','op_year','axil_node','surv_stat'])
```

```python
data.surv_stat.value_counts()
```

```python
sns.FacetGrid(data,hue='surv_stat',height=8).map(sns.distplot,'axil_node').add_legend()
```

```python
sns.set_style("whitegrid");
sns.pairplot(data,hue='surv_stat',height=4)
plt.show()
```

```python
data[data.axil_node==0][data.surv_stat==2]
```

```python
counts, bin_edges = np.histogram(data['axil_node'],  density = True)
pdf = counts/(sum(counts))
print(pdf);
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)

```

```python
sns.violinplot(x=data.surv_stat,y=data.axil_node,data=data)
```
